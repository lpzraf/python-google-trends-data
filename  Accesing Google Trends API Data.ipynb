# Accesing Google Trends API Data

## Libraries

```python
import pandas as pd
from pytrends.request import TrendReq #library for Google Trends
%matplotlib inline
```

```python
pytrends = TrendReq(hl='en-US',
                   tz=360, #timezone
                   timeout=(10,25), # timeout after a few secs
                   retries=2, #retries twice
                   backoff_factor=0.1) # stops after a few attempts
```

```python
kw_list = ['seo audit','seo agency','app marketing agency','technical seo','technical marketing']
```

```python
# To change categories, locations or regions: 
# https://github.com/pat310/google-trends-api/wiki/Google-Trends-Categories    
```

```python
# Pass my var to Google Trends
pytrends.build_payload(kw_list, #keywords to lookup
                       cat=83, #category to lookup
                       timeframe='today 5-y', # timeframe
                       geo='', #country
                       gprop='') #region
```

```python
# Convert GTrends data to a Pandas DataFrame.
trends_data = pytrends.interest_over_time()
```

```python
trends_data.plot()
```

# Visualizing Google Trends

```python
# Plotting a line graph with a secondary y of the largest value
trends_data.plot(secondary_y=('seo agency'), figsize=(15,5))
```

```python
# Plotting a chart of cumulative sum of all the data.
trends_data.cumsum().plot(figsize=(15,5))
```
